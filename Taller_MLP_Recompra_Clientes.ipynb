```python
# BLOQUE 1: SUBIDA DEL ARCHIVO Y CARGA DE DATOS
from google.colab import files
uploaded = files.upload()

import pandas as pd

# Cargar el archivo Excel subido
df = pd.read_excel("dataset_recompra_clientes.xlsx")
df.head()
```

```python
# Informacion general del archivo
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
# BLOQUE 2: ANLISIS INICIAL Y NORMALIZACIN
from sklearn.preprocessing import StandardScaler

# Separar variables predictoras y variable objetivo
X = df.drop("recompra", axis=1)
y = df["recompra"]

# Normalizar variables numricas para mejor desempeo del modelo
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Mostrar una muestra de los datos transformados
pd.DataFrame(X_scaled, columns=X.columns).head()
```

```python
# BLOQUE 3: DIVISIN DE DATOS Y CONSTRUCCIN DEL MODELO
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Dividir en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Construccin del modelo MLP
model = Sequential([
    Dense(16, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')  # Para clasificacin binaria
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenamiento
hist = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.2, verbose=0)
```

```python
# BLOQUE 4: VISUALIZACIN DEL ENTRENAMIENTO
import matplotlib.pyplot as plt

# Precisin
plt.plot(hist.history['accuracy'], label='Entrenamiento')
plt.plot(hist.history['val_accuracy'], label='Validacin')
plt.title('Precisin del Modelo')
plt.xlabel('pocas'); plt.ylabel('Precisin')
plt.legend(); plt.grid(); plt.show()

# Prdida
plt.plot(hist.history['loss'], label='Entrenamiento')
plt.plot(hist.history['val_loss'], label='Validacin')
plt.title('Prdida del Modelo')
plt.xlabel('pocas'); plt.ylabel('Prdida')
plt.legend(); plt.grid(); plt.show()
```

```python
# BLOQUE 5: EVALUACIN DEL MODELO
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt


y_pred_probs = model.predict(X_test)
y_pred = (y_pred_probs > 0.5).astype(int)

# Reporte con precisin, recall y F1-score
print(classification_report(y_test, y_pred, target_names=["No Recompra", "Recompra"]))

#Matriz de confusion
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No Recompra", "Recompra"],
            yticklabels=["No Recompra", "Recompra"])
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.title("Matriz de Confusin")
plt.show()
```

```python
# BLOQUE 6: PREDICCIN DE NUEVOS CASOS
nuevos_clientes = pd.DataFrame({
    "dias_ultima_compra": [15, 250],
    "total_compras": [8, 3],
    "monto_promedio": [120, 30],
    "uso_cupon": [0, 1],
    "tiempo_navegacion": [20, 4],
    "numero_visitas": [5, 1]
})

nuevos_scaled = scaler.transform(nuevos_clientes)
preds_nuevos = model.predict(nuevos_scaled)

nuevos_clientes["Probabilidad_Recompra"] = preds_nuevos
nuevos_clientes["Prediccin"] = (preds_nuevos > 0.5).astype(int)

print("Prediccin de nuevos clientes:")
print(nuevos_clientes)
```

```python
#Barra de probabilidad de recompra para cada cliente
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
bars = plt.bar(['Cliente 1', 'Cliente 2'], nuevos_clientes['Probabilidad_Recompra'], color=['skyblue', 'orange'])
plt.title("Probabilidad de Recompra\n")
plt.ylabel("Probabilidad")
plt.ylim(0,1)

# Muestra el valor encima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.02, f"{yval:.2f}", ha='center', va='bottom')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Selecciona solo las columnas de entrada
columnas_entrada = ["dias_ultima_compra", "total_compras", "monto_promedio", "uso_cupon", "tiempo_navegacion", "numero_visitas"]
clientes_radar = nuevos_clientes[columnas_entrada]

# Normaliza cada variable entre 0 y 1 usando los datos originales (X)
clientes_norm = (clientes_radar - X.min()) / (X.max() - X.min())

# Preparar ngulos para cada variable
labels = clientes_radar.columns
num_vars = len(labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Cerrar el crculo

# Graficar radar
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

for i, row in clientes_norm.iterrows():
    values = row.tolist()
    values += values[:1]  # Cierra la figura
    ax.plot(angles, values, label=f'Cliente {i+1}', marker='o')
    ax.fill(angles, values, alpha=0.15)

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_thetagrids(np.degrees(angles[:-1]), labels)
plt.title("Perfil comparativo de los nuevos clientes\n \n")
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()

```
